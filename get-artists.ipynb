```python
import requests
import pandas as pd
```

```python
access_token = ''
```

```python
headers = {'Authorization': 'Bearer {}'.format(access_token)}
saved_tracks_url = 'https://api.spotify.com/v1/me/tracks'
params = {'limit': 50, 'offset': 4000}

r = requests.get(saved_tracks_url, params=params, headers=headers)
```

```python
r.status_code
```

```python
songs = r.json()
songs['items']
```

```python
songs['items'][0]['track'].keys()
```

```python
songs['items'][0]['track']['artists']
```

```python
set(song['track']['artists'][0]['id']
                     for song in songs['items'])
```

```python
artists = [artist for song in songs['items'] for artist in song['track']['artists']]
```

```python
deduped_artists = set(artist['id'] for artist in artists)
deduped_artists
```

```python
artists_url = 'https://api.spotify.com/v1/artists'
params = {'ids': ','.join(artist for artist in deduped_artists)}

r = requests.get(artists_url, params=params, headers=headers)
```

```python
r.status_code
```

```python
artists = r.json()
```

```python
artists.keys()
```

```python
artists['artists'][0]
```
