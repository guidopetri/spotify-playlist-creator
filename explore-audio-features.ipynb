```python
headers = {'Authorization': 'Bearer {}'.format(access_token)}
saved_tracks_url = 'https://api.spotify.com/v1/me/tracks'
params = {'limit': 50, 'offset': 4000}

r = requests.get(saved_tracks_url, params=params, headers=headers)
```
